```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('Loan.csv')
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.Gender.fillna(df.Gender.index[0],inplace=True)
df.Married.fillna(df.Married.index[0],inplace=True)
df.Dependents.fillna(df.Dependents.index[0],inplace=True)
df.Self_Employed.fillna(df.Self_Employed.index[0], inplace=True)
df.LoanAmount.fillna(df.LoanAmount.mean(), inplace=True)
df.Loan_Amount_Term.fillna(df.Loan_Amount_Term.mean(), inplace=True)
df.Credit_History.fillna(df.Credit_History.median(), inplace=True)
```

```python
df.isnull().sum()
```

```python
df.Gender.unique()
```

```python
df
```

```python
variable= pd.get_dummies(df.Gender, prefix='Gender')
New_Dataset= pd.concat([variable,df], axis='columns')
```

```python
df
```

```python
New_Dataset
```

```python
variable=pd.get_dummies(df.Married, prefix='Married')
New_Dataset=pd.concat([variable,df],axis='columns')
```

```python
New_Dataset
```

```python
del New_Dataset['Loan_ID']
del New_Dataset['Dependents']
```

```python
New_Dataset
```

```python
New_Dataset.drop('Gender', axis='columns', inplace=True)
New_Dataset.drop('Married', axis='columns', inplace=True)
New_Dataset.drop('Education',axis='columns', inplace=True)
```

```python
New_Dataset.drop('Self_Employed',axis='columns', inplace=True)
New_Dataset.drop('Property_Area',axis='columns', inplace=True)
```

```python
New_Dataset
```

```python
X = New_Dataset.iloc[:,:-1]
```

```python
X
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
labelencoder= LabelEncoder()
```

```python
L = labelencoder.fit_transform(New_Dataset['Loan_Status'])
```

```python
L
```

```python
Y = L
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,random_state = 0, test_size = 0.20)
```

```python
import matplotlib.pyplot as plt
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
X.shape
```

```python
Y= Y.reshape
```

```python
acc_list = []
err_list = []
for i in range(1,25):
    model = KNeighborsClassifier(n_neighbors=i)
    model.fit(X_train,Y_train)
    Y_pred = model.predict(X_test)
    acc = accuracy_score(Y_test,Y_pred)
#     print("For k = {}, accuracy = {}".format(i,acc))
    acc_list.append(acc)
    err_list.append(1-acc)
```

```python
plt.plot(list(range(1,25)),err_list,c= 'r')
plt.title('Error rate v/s K')
plt.xlabel('K')
plt.ylabel('Error rate')
plt.show()
```

```python
model = KNeighborsClassifier(n_neighbors=23)
model.fit(X_train,Y_train)
Y_pred = model.predict(X_test)
```

```python

confusion_matrix(Y_test,Y_pred)
```

```python
accuracy_score(Y_test,Y_pred)
```
